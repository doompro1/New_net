```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
import random
import seaborn
from numpy.random import choice
from scipy.stats import *
```

Создайте случайную величину из любого выбранного вами распределения, для разных значений n сгенерируйте 1000 выборок размера n и постройте гистрограммы средних этих выборок.

```python
n=10
p=0.5
rv = sts.norm(n, p)
sample = rv.rvs(size=1000)
```

```python
x = np.linspace(0,2,100)
pdf = rv.pdf(x)
plt.hist(sample, density=True, bins=30, label='sample histogram')
plt.plot(x, pdf, label='theoretical pdf', alpha=0.5)
plt.legend()
plt.ylabel('fraction of samples, $f(x)$')
plt.xlabel('$x$')
plt.show()
```

```python
samples_count = 1000
a = np.array([])
while len(a) != samples_count:
    for number in range(samples_count):
        a = np.append(a , rv.rvs(size= number).mean())
a
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline

mu = 2.0
sigma = 0.5
norm_rv = sts.norm(loc=mu, scale=sigma)
x = np.linspace(-5,7,100)
pdf = norm_rv.pdf(x)
plt.plot(x, pdf)


mu = 2.0
sigma = 2
norm_rv = sts.norm(loc=mu, scale=sigma)
x = np.linspace(-5,7,100)
pdf = norm_rv.pdf(x)
plt.plot(x, pdf)


mu = 2.0
sigma = 1
norm_rv = sts.norm(loc=mu, scale=sigma)
x = np.linspace(-5,7,100)
pdf = norm_rv.pdf(x)
plt.plot(x, pdf)

mu = 2.0
sigma = 0.25
norm_rv = sts.norm(loc=mu, scale=sigma)
x = np.linspace(-5,7,100)
pdf = norm_rv.pdf(x)
plt.plot(x, pdf)

mu = 2.0
sigma =  0.45
norm_rv = sts.norm(loc=mu, scale=sigma)
x = np.linspace(-5,7,100)
pdf = norm_rv.pdf(x)
plt.plot(x, pdf)



lam = 0.1
expon_rv = sts.expon(lam)
expon_rv.rvs(10)
x = np.linspace(-5,7,100)
cdf = expon_rv.cdf(x)
plt.plot(x, cdf)

lam = 1
expon_rv = sts.expon(lam)
expon_rv.rvs(10)
x = np.linspace(-5,7,100)
cdf = expon_rv.cdf(x)
plt.plot(x, cdf)


lam = 3
expon_rv = sts.expon(lam)
expon_rv.rvs(10)
x = np.linspace(-5,7,100)
cdf = expon_rv.cdf(x)
plt.plot(x, cdf)


lam = 0.001
expon_rv = sts.expon(lam)
expon_rv.rvs(10)
x = np.linspace(-5,7,100)
cdf = expon_rv.cdf(x)
plt.plot(x, cdf)

lam = 0.3333333
expon_rv = sts.expon(lam)
expon_rv.rvs(10)
x = np.linspace(-5,7,100)
cdf = expon_rv.cdf(x)
plt.plot(x, cdf)

n = 10
t_rv = sts.t(n)
t_rv.rvs(10)
x = np.linspace(-5,7,100)
cdf = t_rv.cdf(x)
plt.plot(x, cdf)

n = 0.11
t_rv = sts.t(n)
t_rv.rvs(10)
x = np.linspace(-5,7,100)
cdf = t_rv.cdf(x)
plt.plot(x, cdf)

n = 1
t_rv = sts.t(n)
t_rv.rvs(10)
x = np.linspace(-5,7,100)
cdf = t_rv.cdf(x)
plt.plot(x, cdf)

n = 3
t_rv = sts.t(n)
t_rv.rvs(10)
x = np.linspace(-5,7,100)
cdf = t_rv.cdf(x)
plt.plot(x, cdf)
n = 2
t_rv = sts.t(n)
t_rv.rvs(10)
x = np.linspace(-5,7,100)
cdf = t_rv.cdf(x)
plt.plot(x, cdf)


plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.show()
```


Задание 3
Известно, что светофор горит 54 секунды красным, 4 секунды жёлтым и 28 секунд зелёным. Посчитайте, какое количество информации несёт сообщение о цвете светофора в текущий момент. А что будет в случае 28, 28 и 28 секунд для каждого цвета?

```python
import math 
dic = { 'красный' : 54 , 'жёлтый' : 4 , 'зелёный' : 28 }
for color , seconds in dic.items():
    x = sum(dic.values()) * math.log2(seconds)
    print(f"Если передаётся сингал о то, что горит {color}, значит предаётся  {math.ceil(x)} бит информации ")
    
```

```python
import math 
dic = { 'красный' : 28 , 'жёлтый' : 28 , 'зелёный' : 28 }
for color , seconds in dic.items():
    x = math.log2(seconds)
    print(f"Если передаётся сингал о то, что горит {color}, значит предаётся  {math.ceil(x)} бит информации ")
```

Вы работаете секретарем и вам постоянно приходят различные документы. Вы должны быть очень внимательны, чтобы не потерять ни один документ. Каталог документов хранится в следующем виде:

```python
documents = [
    {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
    {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
    {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]
```

Перечень полок, на которых находятся документы хранится в следующем виде:

```python
directories = {
    '1': ['2207 876234', '11-2'],
    '2': ['10006'],
    '3': []
}
```

Необходимо реализовать пользовательские команды (внимание! это не название функций, которые должны быть выразительными, а команды, которые вводит пользователь, чтобы получить необходимый результат):
  
- p – people – команда, которая спросит номер документа и выведет имя человека, которому он принадлежит;  
- s – shelf – команда, которая спросит номер документа и выведет номер полки, на которой он находится  
*Правильно обработайте ситуации, когда пользователь будет вводить несуществующий документ*.
- l – list – команда, которая выведет список всех документов в формате   
**passport "2207 876234" "Василий Гупкин"**  
- as – add_shelf – команда, которая спросит номер новой полки и добавит ее в перечень. *Предусмотрите случай, когда пользователь добавляет полку, которая уже существует*.
- d – delete – команда, которая спросит номер документа и удалит его из каталога и из перечня полок. *Предусмотрите сценарий, когда пользователь вводит несуществующий документ*;  
- m – move – команда, которая спросит номер документа и целевую полку и переместит его с текущей полки на целевую. *Корректно обработайте кейсы, когда пользователь пытается переместить несуществующий документ или переместить документ на несуществующую полку*;  
- a – add – команда, которая добавит новый документ в каталог и в перечень полок, спросив его номер, тип, имя владельца и номер полки, на котором он будет храниться. *Корректно обработайте ситуацию, когда пользователь будет пытаться добавить документ на несуществующую полку*.

```python
documents = [
  {"type": "passport", "number": "2207 876234", "name": "Василий Гупкин"},
  {"type": "invoice", "number": "11-2", "name": "Геннадий Покемонов"},
  {"type": "insurance", "number": "10006", "name": "Аристарх Павлов"}
]

directories = {
  '1': ['2207 876234', '11-2'],
  '2': ['10006'],
  '3': []
}


def p(documents):
  number = ""
  number = input('Введите номер: ')
  name = ""

  for items in documents:
    if items['number'] == number:
      name = items['name']

  if name != '':    
    print()
    print(name)
    print()
  else:
    print()
    print('Номера не существует')
    print()


def s(directories):
  number = ""
  shelf_name = ""
  number = input('Введите номер: ')

  for items, values in directories.items():
    if number in values:
      shelf_name = items
  
  if shelf_name != '':    
    print()
    print(f'{shelf_name} полка')
    print()
  else:
    print()
    print('Номера не существует')
    print()

def l(documents):
  n = 0
  for elements in documents:
    type_ = documents[n]['type']
    num_ = documents[n]['number']
    name_ = documents[n]['name']
    print(f'{type_} "{num_}" "{name_}"')
    n += 1

def a(documents, directories):
  type_ = input('Введите тип документа: ')
  number_ = input('Введите номер документа: ')
  name_ = input('Введите имя: ')
  shelf_ = input('Введите номер полки: ')

  directories_list = []

  for strings in directories:
    directories_list.append(strings)

  check = False

  if shelf_ in directories_list:

    for keys, val in directories.items():
      if keys == shelf_:
        val = val.append(number_)
        check = True
        pass
      
  if check != True:
    print()
    print('Полки не существует')
    print()

  def dict_create(type_, number_, name_):
    new_dict = {}
    new_dict ["type"] = type_
    new_dict ["number"] = number_
    new_dict ["name"] = name_
    return new_dict

  documents.append(dict_create(type_, number_, name_))

def m(directories):
  number = input('Номер документа: ')
  from_ = input('Номер полки, откуда переместить: ')
  to_ = input('Номер полки, куда переместить: ')
  
  some_list = list()

  all_numbers = set()
  
  for items, values in directories.items():
    all_numbers.update(values)
  
  
  if number in all_numbers:

    check = False

    for keys, items in directories.items():
      if keys == from_ and number in items:
        check = True

    if check == True:
      if to_ in directories:
        # print('получилось')

        for keys, values in directories.items():
          if keys == from_:
            some_list = values
            some_list.remove(number)
            directories[keys] = some_list
          
          if keys == to_:
            some_list = values
            some_list.append(number)
            directories[keys] = some_list
            
        
      else:
        print('Полка назначения задана неправильно.')   
    else:
      print('Исходная полка задана неправильно.')
  else:
      print('Документа нет на полке или не существует')

def ads(directories):
  number = input('Номер новой полки: ')

  if number not in directories:
    directories[number] = list()
 

  else: 
    print('Полка существует')

def d(documents, directories):
  number = input('Номер документа: ')
  all_numbers = set()
  
  for items, values in directories.items():
    all_numbers.update(values)
  
  if number in all_numbers:

    for items, values in directories.items():
      for lists in values:
        if number in lists:
          values.remove(number)


    for items in documents:
      if number in items.values():
        documents.remove(items)

  
  else:
    print('Документа не существует')
  


def main(documents, directories):

  command_list = ['p', 's', 'l', 'a', 'd', 'm', 'ass', 'q']

  print(f'Список комманд: {", ".join(command_list)} ')

  while True:
    command = input('Введите команду: ')
    if command == 'p':
      p(documents)
    elif command == 's':
      s(directories)
    elif command == 'l':
      l(documents)
    elif command == 'a':
      a(documents, directories)
      print(documents)
      print(directories)
    elif command == 'm':
      m(directories)
      print(documents)
      print(directories)
    elif command == 'd':
      d(documents, directories)
      print(documents)
      print(directories)
    elif command == 'ads':
      ads(directories)
      print(documents)
      print(directories)
    elif command == 'q': 
      print('До свидания!')
      break
    elif command not in command_list:
      print()
      print('Комманды не существует') 
      print()

main(documents, directories)
```

```python
import pandas as pd
df = pd.read_csv('video_games.csv')
df.head()
```

```python
df1 = df.pivot_table(index = 'Name' , columns = 'Genre' , values = 'Critic_Score')
df1sports = df1.dropna(subset=['Sports'])
### Проверим на выбросы 
table = [x for x in df1sports['Sports'] if x < 10.0  or x > 100.0]
len(table)
```

```python
from scipy import stats as st
alpha = 0.05
result = st.ttest_1samp(df1sports['Sports'], 80)
print(result)
if (result.statistic < 0) & (result.pvalue / 2 < alpha):
    print('Нельзя сказать , что критики в целом оценивают спортивные игры больше 80 из 100')
else:
    print('Нельзя сказать , что критици в целом оценивают спортивные игры меньше  80 ')
```

```python
import glob
import numpy as np
import pandas as pd
import re
import nltk
import gensim
import sklearn
from gensim import corpora
from sklearn.datasets import load_wine
import matplotlib.pyplot as plt
from matplotlib import colors
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.metrics import confusion_matrix
```

```python
data = pd.read_csv('spam.csv')
data.head()
```

```python
data_clear = data.dropna()
message_list = [x for x in data_clear['Message'] ]
categ_list =  [x for x in data_clear['Category'] ]
message_list = [text.lower() for text in message_list]
message_list = [re.sub('[\W_]+',' ', text) for text in message_list]
message_list = [text.split() for text in message_list]
```

```python
from nltk.corpus import stopwords
stopwords_set = set(stopwords.words('english'))
message_list = [[word for word in text if word not in stopwords_set] for text in message_list]
```

```python
from nltk.stem import WordNetLemmatizer
wordnet_lemmatizer = WordNetLemmatizer()
lemmatized_messages = [[wordnet_lemmatizer.lemmatize(word) for word in text] for text in message_list]
```

```python
str_lem_mes = []
for x in lemmatized_messages:
    String = ' '.join(x)
    str_lem_mes.append(String)
data['lem_mes'] = str_lem_mes
```

```python
data['index1'] = data.index
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer()
tfidf_matrix = tfidf.fit_transform(data.lem_mes)
names = tfidf.get_feature_names()
tfidf_matrix = pd.DataFrame(tfidf_matrix.toarray())
tfidf_matrix['Category'] = categ_list
tfidf_matrix['index1'] = tfidf_matrix.index
```

```python
data_train, data_test = train_test_split(tfidf_matrix, test_size=0.3, random_state=42)
```

```python
y_train, y_test, y_index = data_train['Category'], data_test['Category'], data_test['index1']

del data_train['Category']
del data_train['index1']
del data_test['Category']
del data_test['index1']
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
lda = LinearDiscriminantAnalysis()
lda.fit(data_train, y_train)
lda.predict(data_test)
```

# Матрица 

```python
sklearn.metrics.confusion_matrix(y_test.values, lda.predict(data_test) , labels=["ham", "spam"])
```

Для этого потребуется получать страницы статей и искать по тексту внутри этой страницы.

Итоговый датафрейм формировать со столбцами: <дата> - <заголовок> - <ссылка> - <текст_статьи>

```python
Posts_table = pd.DataFrame()
post_id_list = []
post_data_list = []
post_title_list = []
post_link_list = []
keywords_list = []
def POST_ID_with_keywords(posts, x):
    global post_id_list
    global keywords_list
    for post in posts:
        post_id = post.parent.attrs.get('id')
        post_id = int(post_id.split('_')[-1])
        post_id_list.append(post_id)
        keywords_list.append(x)
    if not post_id:
        post_id = 'Нет данных'
        post_id_list.append(post_id)
        keywords_list.append(x)
        
def POST_DATE(posts):
    global post_data_list
    for post in posts:
        post_data = post.find('span', class_='post__time')
        post_data_list.append(post_data.text)
    if not post_data:
        post_data = 'Нет данных'
        post_data_list.append(post_data)
        
def POST_TITLE(posts):
    global post_title_list
    for post in posts:
        post_title = post.find('a', class_='post__title_link')
        post_title_list.append(post_title.text)
    if not post_title:
        post_title = 'Нет данных'
        post_title_list.append(post_title)

def POST_LINK(posts):
    global post_links_list
    for post in posts:
        post_link = post.find('a').get('href')
        post_link_list.append(post_link)
    if not post_link:
        post_link = 'Нет данных'
        post_link_list.append(post_link)
        
        

KEYWORDS = ['python', 'парсинг', 0]
URL = 'https://habr.com/ru/all/'
for x in KEYWORDS :
    if x != 0 :
        params = {
        'search-form__field': x}
        res = requests.get(URL, params)
        params_for_time = {
            'toggle-menu__item-link' : 'по времени'
        }
        res_sorted = requests.get(res.url, params_for_time )
        soup = BeautifulSoup(res_sorted.text, 'html.parser')
        posts = soup.find_all('article', class_='post')
        POST_ID_with_keywords(posts, x)
        POST_DATE(posts)
        POST_TITLE(posts)
        POST_LINK(posts)
list_post = list(zip(post_id_list , keywords_list , post_title_list , post_data_list , post_link_list))
row = 0 
while row < len(list_post) :
    _row_ = {'Post ID' : list_post[row][0] , 'Post keyword': list_post[row][1] , 'Post title': list_post[row][2] , 'Post date' : list_post[row][3] , "Post link" : list_post[row][4]}
    Posts_table = pd.concat([Posts_table, pd.DataFrame([_row_])]) 
    row += 1
print(Posts_table)
```
